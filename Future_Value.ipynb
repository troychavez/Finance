```python
import numpy as np
import matplotlib.pyplot as plt
```

In this exercise, we examine an initial investment of $1,000 over a ten-year period under both monthly and continuous compounding considering a 20% interest rate.

```python
# Parameters:
N = 12
P0 = 1*10**3
r = 20/100
```

```python
# Monthly compounding:
years = 10
t = np.arange(N*years + 1) / (N)
F_ERA = P0 * np.exp(r * t)
```

```python
# Continuous compounding:
t_d = np.arange((N*years)+1)
F_d = P0 * (1 + r/N) ** t_d
```

```python
plt.figure(figsize = (12, 8))
plt.step(t_d, F_d, where='post', label = 'Monthly Compounded')
plt.plot(t_d, F_ERA, label = 'Continuously Compounded')
plt.grid()
plt.title('Compound Interest')
plt.xlabel('t [Months]')
plt.ylabel('Investment [$]')
plt.legend()
plt.show()
```
